```python
empreendimento = pd.read_csv(r'../../data/Empreendimento.csv', sep=';', encoding='ISO-8859-1')
cod_empreendimento =  empreendimento[empreendimento['PRODUTO'] == 'SOJA']
cod_empreendimento
```

```python
ref_bacen = sicor_basico[sicor_basico['CD_EMPREENDIMENTO'].isin(set(cod_empreendimento['CODIGO']))]
ref_bacen
```

```python
glebas = pd.read_csv(r'C:/Users/leall/OneDrive/Área de Trabalho/03_TABS_COMP_BASICAS_OPERACOES_CREDITO_RURAL_PROAGRO_RECURSOS_PUB/Glebas.csv', sep=';', encoding='ISO-8859-1')
glebas_soja = glebas[(glebas['REF_BACEN'].isin(set(ref_bacen['REF_BACEN']))) & (glebas['NU_ORDEM'].isin(set(ref_bacen['NU_ORDEM'])))]
glebas_soja
```

```python
# Suponha que 'count_registers' seja uma série que contém contagens para cada valor único em 'REF_BACEN'.
# Primeiro, crie uma série com as contagens dos valores 'REF_BACEN'.
count_registers = glebas_soja['REF_BACEN'].value_counts()

# Em seguida, crie uma lista de valores 'REF_BACEN' que têm menos de 3 registros.
codigos_a_remover = count_registers[count_registers <= 3].index.tolist()

# Finalmente, filtre o DataFrame 'glebas_soja' para manter apenas os registros em que 'REF_BACEN' não está na lista de códigos a serem removidos.
final = glebas_soja[~glebas_soja['REF_BACEN'].isin(codigos_a_remover)]
final.to_csv("filtradoteste.csv")
```
